```python
import numpy as np
import matplotlib.pyplot as plt

x0 = np.linspace(0, 10, 100)
y1 = 2 + 2 * np.sin(2 * x0)

fig, axes = plt.subplots(1, 2, figsize=(10, 4))  # 1行2列の図を作成

# サブプロットの1つ目にsin関数のグラフを描画
axes[0].plot(x0, y1, linewidth=2.0)
axes[0].set_xlim(1, 10)
axes[0].set_xticks(np.arange(0, 11, 2))
axes[0].set_ylim(-1, 6)
axes[0].set_yticks(np.arange(-1, 7, 1))
axes[0].set_title('sin関数のグラフ')

plt.tight_layout()  # グラフのレイアウトを整える
plt.show()
```
